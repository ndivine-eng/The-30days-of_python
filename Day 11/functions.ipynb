```python
import math
from collections import Counter

# 1. Function to add two numbers
def add_two_numbers(a, b):
    return a + b

# 2. Function to calculate area of a circle
def area_of_circle(r):
    return math.pi * r ** 2  # Ï€ * r^2

# 3. Function to add all numbers (with type checking)
def add_all_nums(*args):
    if all(isinstance(i, (int, float)) for i in args):
        return sum(args)
    return "Error: All arguments must be numbers."

# 4. Function to convert Celsius to Fahrenheit
def convert_celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# 5. Function to check season based on month
def check_season(month):
    month = month.lower()
    seasons = {
        "winter": ["december", "january", "february"],
        "spring": ["march", "april", "may"],
        "summer": ["june", "july", "august"],
        "autumn": ["september", "october", "november"]
    }
    for season, months in seasons.items():
        if month in months:
            return season.capitalize()
    return "Invalid month"

# 6. Function to calculate slope of a line
def calculate_slope(x1, y1, x2, y2):
    return "Undefined (vertical line)" if x1 == x2 else (y2 - y1) / (x2 - x1)

# 7. Function to solve a quadratic equation
def solve_quadratic_eqn(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "No real solution"
    x1 = (-b + math.sqrt(discriminant)) / (2 * a)
    x2 = (-b - math.sqrt(discriminant)) / (2 * a)
    return x1 if discriminant == 0 else (x1, x2)

# 8. Function to print list
def print_list(lst):
    for item in lst:
        print(item)

# 9. Function to reverse a list
def reverse_list(lst):
    return lst[::-1]

# 10. Function to capitalize list items
def capitalize_list_items(lst):
    return [item.capitalize() for item in lst]

# 11. Function to add an item to a list
def add_item(lst, item):
    return lst + [item]

# 12. Function to remove an item from a list
def remove_item(lst, item):
    return [i for i in lst if i != item]

# 13. Function to sum numbers in a range
def sum_of_numbers(n):
    return n * (n + 1) // 2

# 14. Function to sum odd numbers in a range
def sum_of_odds(n):
    return sum(i for i in range(n + 1) if i % 2 != 0)

# 15. Function to sum even numbers in a range
def sum_of_even(n):
    return sum(i for i in range(n + 1) if i % 2 == 0)

# 16. Function to count even and odd numbers
def evens_and_odds(n):
    evens = n // 2 + 1
    odds = n - evens + 1
    return f"The number of odds: {odds}, The number of evens: {evens}"

# 17. Function to compute factorial
def factorial(n):
    return math.factorial(n)

# 18. Function to check if a value is empty
def is_empty(param):
    return not bool(param)

# 19. Function to calculate mean
def calculate_mean(lst):
    return sum(lst) / len(lst)

# 20. Function to calculate median
def calculate_median(lst):
    lst.sort()
    mid = len(lst) // 2
    return lst[mid] if len(lst) % 2 != 0 else (lst[mid - 1] + lst[mid]) / 2

# 21. Function to calculate mode
def calculate_mode(lst):
    counter = Counter(lst)
    max_count = max(counter.values())
    return [k for k, v in counter.items() if v == max_count]

# 22. Function to calculate range
def calculate_range(lst):
    return max(lst) - min(lst)

# 23. Function to calculate variance
def calculate_variance(lst):
    mean = calculate_mean(lst)
    return sum((x - mean) ** 2 for x in lst) / len(lst)

# 24. Function to calculate standard deviation
def calculate_std(lst):
    return math.sqrt(calculate_variance(lst))

# 25. Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# 26. Function to check if all elements in a list are unique
def all_unique(lst):
    return len(lst) == len(set(lst))

# 27. Function to check if all items in a list are of the same type
def all_same_type(lst):
    return all(isinstance(item, type(lst[0])) for item in lst)

# 28. Function to check if a variable name is valid in Python
def is_valid_variable(var):
    return var.isidentifier()

# 29. Function to find the most spoken languages
def most_spoken_languages(data, top_n=10):
    languages = Counter(lang for country in data for lang in country['languages'])
    return languages.most_common(top_n)

# 30. Function to find the most populated countries
def most_populated_countries(data, top_n=10):
    return sorted(data, key=lambda x: x['population'], reverse=True)[:top_n]
```
